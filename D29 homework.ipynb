### Q1:計算標準常態分配，小於1的機率有多大?
![](Q1.png)

```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
## Q1
#  提示: 用 cdf 來算
mu=0
sigma=1
x=np.linspace(-1, 1, 500)
y = stats.norm.pdf(x,mu, sigma)
plt.figure('A')
plt.plot(x, y, "b-", linewidth=2)

cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.figure('B')
plt.plot(x, cumsum_probs,"b-", linewidth=2)

x_index = np.where(x==1)
print(cumsum_probs[x_index])
```

### Q2:計算標準常態分配，大於1，小於 -1 的機率有多大?
![](Q2.png)


```python
## Q2
'''
你的程式碼
'''

print(1-cumsum_probs[1])
x_index2 = np.where(x==-1)
print(cumsum_probs[x_index2])
```

### Q3:X~N(2,4),x 服從常態分配，平均數為2,變異數為 4，計算 X小於 6 的機率有多大?

```python
## Q3
#  提示: 和 Q1,Q2 不同，要先標準化。
'''
你的程式碼
'''

mu=2
sigma=4
x=np.linspace(-4,6, 500)
y = stats.norm.pdf(x,mu, sigma)
plt.figure('A')
plt.plot(x, y, "b-", linewidth=2)

cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.figure('B')
plt.plot(x, cumsum_probs,"b-", linewidth=2)

print(cumsum_probs[6])
```
