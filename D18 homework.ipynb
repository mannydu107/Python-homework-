# 範例重點

如何使用亂數, 資料集來操作


```python
# 載入需要的...
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 360)
y = np.cos(x*2*np.pi/360)
plt.plot(x, y)
plt.xlim([0, 360])
plt.xticks([0, 45, 90, 135, 180, 225, 270, 315, 360])
plt.savefig("D18 homework.png",dpi=300,format="png")
```

# 散點圖: Scatter Plots
    
    顏色由（X，Y）的角度給出。

    注意標記的大小，顏色和透明度。

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)


angle = np.arctan(Y/X)
plt.scatter(X, Y, c=angle, alpha=0.3)



plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.show()
```
